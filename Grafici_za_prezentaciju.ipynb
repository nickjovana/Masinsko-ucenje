```python
    import matplotlib.pyplot as plt
    import numpy as np
    feature_names=['KNN-17', 'Basjes-0.01','OnevsOne','OnevsRest']
    plt.figure(figsize = (10, 5))
    plt.bar(np.arange(0, 4),[0.67,0.68,0.69,0.74])
    plt.xticks(np.arange(0, 4), feature_names, rotation='horizontal')
    plt.title('Rezultati za kompletan skup za najbolje parametre')
    plt.show()
```
